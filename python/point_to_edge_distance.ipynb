```python
def case_1(a, b, c):
    ac = c - a
    return ac.dot(ac)

def case_2(a, b, c):
    bc = c - b
    return bc.dot(bc)

def case_3(a, b, c):
    ab = b - a
    ac = c - a
    bc = c - b
    e = ac.dot(ab)
    f = ab.dot(ab)
    return ac.dot(ac) - e * e / f
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([-1.0, 0.0])
b = np.array([1.0, 0.0])
```

```python
def eval_case(case_, minv, maxv):
    x = np.linspace(minv*10.0,maxv*10.0)/10.0
    y = np.linspace(-10, 10)/10.0
    X, Y = np.meshgrid(x,y)
    xi = X.flatten()
    yi = Y.flatten()
    ci = np.array([xi,yi]).T
    
    z_case_ = np.array([case_(a,b,c) for c in ci])
    z_case_ = np.reshape(z_case_, newshape=X.shape)
    return z_case_, X, Y
```

```python
z_case_1, X1, Y1 = eval_case(case_1, -2, -0.8)
z_case_2, X2, Y2 = eval_case(case_2, 0.8, 2)
z_case_3, X3, Y3 = eval_case(case_3,  -2, 2)
```
